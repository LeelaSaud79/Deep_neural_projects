```python
import pandas_datareader as pdr
```

```python
df=pdr.DataReader('AMZN','yahoo',start='2000-01-01',end='2022-01-01')
```

```python
df.head()
```

```python
trainrows=df.iloc[:4000]
testrows=df.iloc[4000:]
```

```python
window_size=50
trainx=[ ]
trainy=[ ]
```

```python
trainrows=trainrows['Close'].tolist()
trainrows
```

```python
for i in range(len(trainrows)-window_size):
    trainx.append(trainrows[i:i+window_size])
    trainy.append(trainrows[i+window_size])
```

```python
import numpy as np
trainx=np.array(trainx)
```

```python
trainx
```

```python
trainy
```

```python
testrows=testrows['Close'].tolist()
testrows
```

```python
testx=[]
testy=[]
for i in range(len(testrows)-window_size):
    testx.append(testrows[i:i+window_size])
    testy.append(testrows[i+window_size])
testx=np.array(testx)
```

```python
from tensorflow.keras.layers import Input,Dense,SimpleRNN,LSTM,GRU
from tensorflow.keras import Sequential
from tensorflow.keras.losses import MeanAbsoluteError
from tensorflow.keras.optimizers import SGD
```

```python
model=Sequential([
    LSTM(10,input_shape=(window_size,1),return_sequences=True),
    LSTM(10,),
    Dense(100,activation='relu'),
    Dense(1, activation='relu')
])
```

```python
model.summary()
```

```python
train=model.compile(loss=MeanAbsoluteError(),optimizer=SGD(),metrics=['mae'])
```

```python
model.fit(trainx.tolist(),trainy,epochs=100,batch_size=64)
```
