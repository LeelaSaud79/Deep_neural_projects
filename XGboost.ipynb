```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('me_data.csv')
data.head()
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data = data.dropna()
data
```

```python
data.shape
```

```python
data.info()
```

Selectiong only the columns we are about to use:

```python
useful_cols = ['Rooms', 'Distance', 'Landsize', 'BuildingArea', 'YearBuilt']
useful_cols
```

```python
x= data[useful_cols]
y= data.Price
```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
```

```python
xtrain, xtest, ytrain, ytest =  train_test_split(x,y,test_size = 0.3, random_state = 32)
xtrain
```

```python
model = XGBRegressor()
model
```

```python
model.fit(xtrain, ytrain)
```

Now predicting and evalauating the model:

```python
from sklearn.metrics import mean_absolute_error
predict = model.predict(xtest)
```

```python
print("Mean Absolute Error: " + str(mean_absolute_error(predict, ytest)))
```

```python
list(zip(ytest, predict))
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
model = XGBRegressor(n_estimators = 600)
model.fit(xtrain, ytrain,
         early_stopping_rounds = 5,
         eval_set = [(xtest, ytest)],
         verbose= False)
# early_stopping_rounds provides a way to automatically find the ideal value for n_estimators.
```

```python
model = XGBRegressor(n_estimators = 600, learning_rate = 0.06)
model.fit(xtrain, ytrain,
         early_stopping_rounds = 5,
         eval_set = [(xtest, ytest)],
         verbose= False)
```

```python
model = XGBRegressor(n_estimators = 600, learning_rate = 0.06, n_jobs = 5)
model.fit(xtrain, ytrain,
         early_stopping_rounds = 5,
         eval_set = [(xtest, ytest)],
         verbose= False)
```
